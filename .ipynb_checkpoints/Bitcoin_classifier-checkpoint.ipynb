```python
# Define the X and Y columns and train Random Forest


import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    classification_report, confusion_matrix,
    roc_auc_score, average_precision_score,
    precision_recall_curve
)

# 1. Load the normalized dataset
df = pd.read_csv("df_8_vars_normalized.csv")
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# 2. Define features and label
feature_cols = ["open", "high", "low", "close", "volume", "ret_1d", "ret_3d", "ret_5d"]
label_col = "trigger"

# 3. Remove NaNs in features or label
mask = df[feature_cols + [label_col]].notna().all(axis=1)
df = df.loc[mask]

# 4. Time-based split (80/20)
cut = int(len(df) * 0.8)
X_train, X_test = df[feature_cols].iloc[:cut], df[feature_cols].iloc[cut:]
y_train, y_test = df[label_col].iloc[:cut], df[label_col].iloc[cut:]

print(f"✅ Dataset shape: {df.shape}")
print(f"Train: {X_train.shape},  Test: {X_test.shape}")
print(f"Train period: {df['date'].iloc[0].date()} → {df['date'].iloc[cut-1].date()}")
print(f"Test  period: {df['date'].iloc[cut].date()} → {df['date'].iloc[-1].date()}")

# 5. Fit Random Forest
rf = RandomForestClassifier(
    n_estimators=800,
    max_depth=8,
    min_samples_leaf=20,
    class_weight={0:1, 1:3},  # emphasize minority class a bit more
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train, y_train)

# 6. Predictions
proba = rf.predict_proba(X_test)[:, 1]
y_pred = (proba >= 0.5).astype(int)

# 7. Reports
print("\n=== RANDOM FOREST REPORT (threshold = 0.5) ===")
if y_test.nunique() > 1:
    print("ROC-AUC:", roc_auc_score(y_test, proba))
else:
    print("ROC-AUC: skipped (test set has only one class)")
print("PR-AUC:", average_precision_score(y_test, proba))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, digits=3))

# 8. (Optional) Find threshold that balances precision/recall
prec, rec, thr = precision_recall_curve(y_test, proba)
target_prec = 0.55
cand = np.where(prec[:-1] >= target_prec)[0]
idx = cand[np.argmax(rec[cand])] if len(cand) else np.argmax(rec)
best_thr = thr[idx] if len(thr) else 0.5

y_pred_adj = (proba >= best_thr).astype(int)
print(f"\n=== Adjusted Threshold (thr={best_thr:.3f}, target precision≥{target_prec}) ===")
print(f"Precision: {prec[idx]:.3f}   Recall: {rec[idx]:.3f}")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_adj))
print("\nClassification Report:\n", classification_report(y_test, y_pred_adj, digits=3))
```

```python
# Parse date (common cases)
if "date" in df.columns:
    df["date"] = pd.to_datetime(df["date"], errors="coerce")
elif "open time" in df.columns:
    # often like "2020-01-01 00:00:00 UTC"
    df["date"] = pd.to_datetime(df["open time"].str.replace(" UTC","", regex=False), errors="coerce", utc=True)
else:
    raise ValueError("Could not find a date column. Make sure CSV has 'date' or 'Open time'.")

# Ensure required OHLCV columns exist (rename if needed)
rename_map = {
    "open": "open", "high": "high", "low": "low", "close": "close", "volume": "volume"
}
for k in list(rename_map):
    if rename_map[k] not in df.columns:
        # try capitalized version
        if k.capitalize() in df.columns:
            df.rename(columns={k.capitalize(): k}, inplace=True)
        elif k.upper() in df.columns:
            df.rename(columns={k.upper(): k}, inplace=True)

df = df.dropna(subset=["date","open","high","low","close"]).sort_values("date").reset_index(drop=True)
df.head()
```

```python
def build_features_and_labels(data, horizon=6, gain=0.06):
    df = data.copy()

    # Daily returns / momentum
    df["ret_1d"] = df["close"].pct_change()
    df["mom_5d"] = df["close"].pct_change(5)

    # Rolling 20-day stats (you can add 60/100 later)
    win = 20
    df["ma20_close"] = df["close"].rolling(win).mean()
    df["ma20_vol"]   = df["volume"].rolling(win).mean()
    df["hhv20"]      = df["high"].rolling(win).max()
    df["llv20"]      = df["low"].rolling(win).min()

    # True Range & ATR14
    tr = pd.concat([
        (df["high"] - df["low"]),
        (df["high"] - df["close"].shift()).abs(),
        (df["low"]  - df["close"].shift()).abs()
    ], axis=1).max(axis=1)
    df["atr14"] = tr.rolling(14).mean()

    # Features
    rng = (df["hhv20"] - df["llv20"]).replace(0, np.nan)
    df["close_ma20_pct"] = df["close"]/df["ma20_close"] - 1
    df["pos_20d"]        = (df["close"] - df["llv20"])/rng
    df["dist_hi_20d"]    = df["close"]/df["hhv20"] - 1
    df["dist_lo_20d"]    = df["close"]/df["llv20"] - 1
    df["atr14_norm"]     = df["atr14"]/df["close"]
    df["ret_std20"]      = df["ret_1d"].rolling(win).std()
    df["vol_ma20_pct"]   = df["volume"]/df["ma20_vol"] - 1
    df["vol_z20"]        = (df["volume"] - df["volume"].rolling(win).mean()) / df["volume"].rolling(win).std()
    df["body_pct"]       = (df["close"] - df["open"])/df["open"]
    df["range_pct"]      = (df["high"] - df["low"])/df["close"]

    # Label: close->open entry; +6% first-touch by High within 'horizon' days
    entry = df["open"].shift(-1)           # next day's open
    tp = entry * (1 + gain)

    hits, days_to = [], []
    H = horizon
    for i in range(len(df)):
        hi = df["high"].iloc[i+1:i+1+H].to_numpy()
        if hi.size == 0:
            hits.append(0); days_to.append(np.nan); continue
        k = next((j for j, v in enumerate(hi, start=1) if v >= tp.iloc[i]), None)
        hits.append(1 if k is not None else 0)
        days_to.append(k if k is not None else np.nan)

    df["trigger"] = hits
    df["days_to_6pct"] = days_to

    # Drop warmup rows with NaNs
    feature_cols = [
        "ret_1d","mom_5d","close_ma20_pct","pos_20d","dist_hi_20d","dist_lo_20d",
        "atr14_norm","ret_std20","vol_ma20_pct","vol_z20","body_pct","range_pct"
    ]
    keep = ["date","open","high","low","close","volume"] + feature_cols + ["trigger","days_to_6pct"]
    out = df[keep].dropna().reset_index(drop=True)
    return out, feature_cols

features, FEATURE_COLS = build_features_and_labels(df, horizon=6, gain=0.06)
features.head()
```

```python
# Train/Validation split by time (simple cut at 80%)
cut = int(len(features)*0.8)
train, test = features.iloc[:cut], features.iloc[cut:]

X_train = train[FEATURE_COLS]
y_train = train["trigger"].astype(int)
X_test  = test[FEATURE_COLS]
y_test  = test["trigger"].astype(int)

rf = RandomForestClassifier(
    n_estimators=600, max_depth=8, min_samples_leaf=80,
    class_weight='balanced', random_state=42, n_jobs=-1
)
rf.fit(X_train, y_train)
p = rf.predict_proba(X_test)[:,1]
print("AUC:", roc_auc_score(y_test, p))
print(classification_report(y_test, (p>=0.65).astype(int)))
```
