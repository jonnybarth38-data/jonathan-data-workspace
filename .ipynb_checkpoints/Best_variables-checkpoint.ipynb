```python
# Load pandas and pathlib
#  Load the 6 vars dataset and verify shape

import pandas as pd
from pathlib import Path

BASE = Path("df_6_vars.csv")  # contains at least: date, open, high, low, close, volume, ret_1d, trigger

df_base = pd.read_csv(BASE)
df_base["date"] = pd.to_datetime(df_base["date"], errors="coerce")

print("Columns:", df_base.columns.tolist())
print("Trigger counts (unchanged check):")
print(df_base["trigger"].value_counts(dropna=False))
df_base.head()
```

```python
# Add vol_1D volitility 1-D

df7 = df_base.copy()
df7["vol_5d"] = df7["ret_1d"].rolling(5).std()

print(df7.columns)
print(df7["trigger"].value_counts(dropna=False))  # Check trigger unchanged
```

```python
# Add rsi_14, which is rsi 14 day

df8 = df7.copy()

delta = df8["close"].diff()
gain = delta.clip(lower=0).rolling(14).mean()
loss = (-delta.clip(upper=0)).rolling(14).mean()
rs = gain / loss
df8["rsi_14"] = 100 - (100 / (1 + rs))

print(df8.columns)
print(df8["trigger"].value_counts(dropna=False))
```

```python
# Add atr_14

df9 = df8.copy()

high_low = (df9["high"] - df9["low"]).abs()
high_prev = (df9["high"] - df9["close"].shift(1)).abs()
low_prev = (df9["low"] - df9["close"].shift(1)).abs()

true_range = pd.concat([high_low, high_prev, low_prev], axis=1).max(axis=1)
df9["atr_14"] = true_range.rolling(14).mean()

print(df9.columns)
print(df9["trigger"].value_counts(dropna=False))
```

```python
# add bollinger band percent

df10 = df9.copy()

sma20 = df10["close"].rolling(20).mean()
std20 = df10["close"].rolling(20).std()

upper = sma20 + 2 * std20
lower = sma20 - 2 * std20

df10["bb_percent"] = (df10["close"] - lower) / (upper - lower)

print(df10.columns)
print(df10["trigger"].value_counts(dropna=False))
```

```python
# add MACD

df11 = df10.copy()

ema12 = df11["close"].ewm(span=12, adjust=False).mean()
ema26 = df11["close"].ewm(span=26, adjust=False).mean()
df11["macd"] = ema12 - ema26

print(df11.columns)
print(df11["trigger"].value_counts(dropna=False))
```

```python
# add MACD signal

df12 = df11.copy()
df12["macd_signal"] = df12["macd"].ewm(span=9, adjust=False).mean()

print(df12.columns)
print(df12["trigger"].value_counts(dropna=False))
```

```python
# Save the added features in the dataset df12 as df_15.csv

df12.to_csv("df_15.csv", index=False)
print("✅ Saved as df_15.csv")
```

```python
# Find the VIF correlation of the variables

import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
import seaborn as sns
import matplotlib.pyplot as plt

# Load your new engineered dataset
df = pd.read_csv("df_15.csv")

# Identify all numeric columns except date + target
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
feature_cols = [col for col in numeric_cols if col != "trigger"]  # keep all features except label

# Drop NaNs (only for correlation & VIF calculation)
df_features = df[feature_cols].dropna()

print("✅ Features being analyzed:", feature_cols)

# 1️⃣ Correlation Matrix (features vs features)
plt.figure(figsize=(12, 8))
sns.heatmap(df_features.corr(), annot=False, cmap="coolwarm")
plt.title("📊 Correlation Matrix of All Engineered Features")
plt.show()

# 2️⃣ Variance Inflation Factor (multicollinearity check)
vif_data = pd.DataFrame({
    "Feature": feature_cols,
    "VIF": [variance_inflation_factor(df_features.values, i) for i in range(len(feature_cols))]
})

print("\n🔥 Variance Inflation Factor (VIF):")
print(vif_data.sort_values("VIF", ascending=False))

```

```python
# Drop the highly correlated variables and save in df_features_reduced.csv
# Starting from df12 or df_15.csv we are keeping in:
#"close", "volume","ret_1d", "ret_3d", "ret_5d","vol_5d", "rsi_14", "atr_14"


import pandas as pd

# 1) Load
df = pd.read_csv("df_15.csv")
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# 2) Add multi-day returns if missing
if "ret_3d" not in df.columns:
    df["ret_3d"] = df["close"].pct_change(3)
if "ret_5d" not in df.columns:
    df["ret_5d"] = df["close"].pct_change(5)

# 3) Select reduced, low-collinearity features
selected_features = [
    "close", "volume",
    "ret_1d", "ret_3d", "ret_5d",
    "vol_5d", "rsi_14", "atr_14"
]

# keep only available columns in case any are missing
available = ["date", "trigger"] + [c for c in selected_features if c in df.columns]

# 4) Drop rows with NaNs (from rolling windows/returns) and save
df_filtered = df[available].dropna().copy()
df_filtered.to_csv("df_features_reduced.csv", index=False)

print("✅ Cleaned & saved as df_features_reduced.csv")
print("Columns used:", available)
print("Shape:", df_filtered.shape)

# --- Optional (recommended before modeling): 1-day lag to avoid leakage ---
# for c in [c for c in df_filtered.columns if c not in ["date","trigger"]]:
#     df_filtered[c] = df_filtered[c].shift(1)
# df_filtered = df_filtered.dropna()
# df_filtered.to_csv("df_features_reduced_lag1.csv", index=False)
# print("✅ Also saved leakage-safe version as df_features_reduced_lag1.csv")
```
